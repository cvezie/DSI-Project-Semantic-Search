```python
client = pymongo.MongoClient('34.215.225.199', 27016)
db_ref = client.wiki_database
wiki_ref = db_ref.wiki_database
```

```python
wiki_data = []
cursor = wiki_ref.find({})
for document in cursor:
    wiki_data.append(document)
```

```python
wiki_df = pd.DataFrame(wiki_data)
wiki_df.head(10)
```

```python
def tfidf_data(df, target, query):
    """
        Params:
        --------
            df: full DataFrame
            target: series, target column from DataFrame
            query: str, 
        
        Returns:
        --------
            something
    """
    tfidf                   = TfidfVectorizer(min_df=5, ngram_range=(1,3))
    X_tf                    = tfidf.fit_transform(target)
    doc_term_matrix_df      = pd.DataFrame(X_tf.toarray(),
                                           index=df.index,
                                           columns=tfidf.get_feature_names())
    
    query                   = pd.Series(query)
    search_terms_encoded    = tfidf.transform(query)
    search_terms_encoded_df = pd.DataFrame(search_terms_encoded.toarray(),
                                           index=query,
                                           columns=tfidf.get_feature_names())
    
    return X_tf, doc_term_matrix_df, search_terms_encoded_df 
```

```python
tfidf_data(wiki_df, wiki_df['text'], "Oracle ERP")
```

```python
def svd_data(df, target, query):
    """
        Params:
        --------
            df:
            target: series
            query: str
        
        Returns:
        --------
            something
    """
    X_tf, doc_term_matrix_df, search_terms_encoded_df = tfidf_data(df, target, query)
                                                                  
    n_components    = 100
    svd             = TruncatedSVD(n_components)
    component_names = ['component_'+str(i+1) for i in range(n_components)]
    
    svd_matrix        = svd.fit_transform(X_tf.toarray())
    svd_search_matrix = svd.transform(search_terms_encoded_df)
    svd_df            = pd.DataFrame(svd_matrix,
                                     index=df.index,
                                     columns=component_names)
    
    return svd_matrix, svd_search_matrix, svd_df, component_names
```

```python
svd_data(wiki_df, wiki_df['text'], "Oracle ERP")
```

```python
def find_page_matches(df, target, query):
    """
        Params:
        --------
            df:
            target: series
            query: str
        
        Returns:
        --------
            something
    """

#     use lsa and cosine similarity to find the most similar vector
    
    svd_matrix, svd_search_matrix, svd_df, component_names = svd_data(df, target, query)
    X_tf, doc_term_matrix_df, search_terms_encoded_df      = tfidf_data(df, target, query)
    
    lsa = pd.DataFrame(svd_matrix,
                       index=doc_term_matrix_df.index,
                       columns=component_names)
    
    lsa['text']       = df.text
    lsa['category']   = df.category
    lsa['pageid']     = df.pageid
    lsa['cosine_sim'] = cosine_similarity(svd_df, svd_search_matrix)
    
#     find the top 5 articles
    top_5_rows = lsa.sort_values('cosine_sim', ascending=False).head(5)
    top_5_articles = top_5_rows[['text', 'category', 'pageid', 'cosine_sim']]
    return top_5_articles
```

```python
find_page_matches(wiki_df, wiki_df['text'], "Oracle ERP")
```
